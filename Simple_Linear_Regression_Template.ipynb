# Simple Linear Regression

## Importing Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Data Set

```
# prompt: import data set

dataset = pd.read_csv('Salary_exp.csv')
X = dataset.iloc[:, :-1].values
#first column
y = dataset.iloc[:, -1].values
#second column
```

## Splitting into test and training set

```
# prompt: import sklearn library  and split the data in test and training set after also explain the code

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# X_train: This is the training set for the independent variable (the first column of the dataset).
# X_test: This is the test set for the independent variable.
# y_train: This is the training set for the dependent variable (the second column of the dataset).
# y_test: This is the test set for the dependent variable.
# test_size: This parameter specifies the proportion of the dataset to include in the test set. In this case, 20% of the data will be used for testing.
# random_state: This parameter ensures that the splitting of the data is reproducible. This means that if you run the code multiple times, you will get the same training and test sets.
```

## Training on Training Set using Simple Linear Regression

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

## Prediction on Test Set

```
y_pred = model.predict(X_test)
```

## Visualizing the Training Set

```
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title("CTC vs years of experience")
plt.xlabel("Experience in years")
plt.ylabel("CTC")
plt.show()
```

## Visualizing the Test Set

```
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title("CTC vs years of experience")
plt.xlabel("Experience in years")
plt.ylabel("CTC")
plt.show()
```
